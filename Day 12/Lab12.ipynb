```python
import tensorflow as tf
import numpy as np
```

# Generate 10 random values for x and y

```python
np.random.seed(42)
x = np.array([np.random.randint(0, 100, 10),np.random.randint(0, 100, 10)])
```

```python
x
```

```python
x = x.reshape(10, 2)
```

```python
x.shape
```

```python
x[:, 0]
```

# Generate z for the above values

```python
z = 5 * x[:, 0] - 3 * x[:, 1] + 7
```

```python
z
```

```python
z.shape
```

# Create a model with 2 inputs to accept x and y (x) and predict z

```python
model = tf.keras.Sequential(
    tf.keras.layers.Dense(units=1, input_shape=[2])
)
```

```python
model.compile(
    loss='mean_squared_error',
    optimizer=tf.keras.optimizers.Adam(0.1)
)
```

# Model summary

```python
model.summary()
```

```python
hist = model.fit(x, z, epochs=1000)
```

```python
hist.history['loss']
```

```python
%matplotlib
import matplotlib.pyplot as plt
```

```python
plt.plot('Epochs', 'Loss', data=hist.history['loss'])
```

```python
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.plot(hist.history['loss']);
```
