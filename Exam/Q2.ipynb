# apply on the following image the shown filter. image link:

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img  = plt.imread('blur.png')
plt.imshow(img)
```

```python
ker  = np.array([[0, -1, 0],
        [-1, 5, -1],
        [0, -1, 0]])
res = cv2.filter2D(img, -1, ker)
plt.imshow(res)
```
