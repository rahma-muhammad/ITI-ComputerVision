# Q1 Make 100*100 rgb pixel image with white background and make diagonal of image with black color.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = np.zeros((256, 256,3),np.uint8)
img1[:,:] = [255, 255, 255]

for i, row in enumerate(img1):
    for j, col in enumerate(row):
        if i == j:
            img1[i ,j]=[0, 0, 0]
            
plt.imshow(img1)
```

# Q2- run video stream using your laptop camera and save frames when i press key “s” close camera when i press “q” ,name of images should be “img.png +{counter}”


```python
cap=cv2.VideoCapture(0)
counter = 0
while (cap.isOpened()):
    state,frame=cap.read()
    if state==True:
        cv2.imshow('Frame',frame)
        k=cv2.waitKey(100)
        if k== ord('q'):
            break
        elif k== ord('s'):
            cv2.imwrite(f'img{counter}.jpg',frame)
            counter += 1
    else:
        break
cap.release()
cv2.destroyAllWindows()
```

# Q3 Make 100*100 rgb pixel image with black background and make diagonal of image with (x) color such that x is came from rgb trackbar

```python
def nothing(x):
    pass

cv2.namedWindow('color')
cv2.createTrackbar('r', 'color', 10, 255, nothing)
cv2.createTrackbar('g', 'color', 10, 255, nothing)
cv2.createTrackbar('b', 'color', 10, 255, nothing)

img=np.zeros((256,256,3), np.uint8)

while 1:
    r = cv2.getTrackbarPos('r', 'color')
    g = cv2.getTrackbarPos('g', 'color')
    b = cv2.getTrackbarPos('b', 'color')
   
    for i, row in enumerate(img):
        for j, col in enumerate(row):
            if i == j:
                img[i ,j]= [b,g,r]
                
    cv2.imshow('color', img)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```

# Q4 create 100*100 rgb image with black background and make 1 vertical line and 1 horizontal line the color of both should be controlled spereat from 2 trackbar First one “horz” and second one is “vert” 


```python
def nothing(x):
    pass

cv2.namedWindow('color')
cv2.createTrackbar('Vert_r', 'color', 0, 255, nothing)
cv2.createTrackbar('Vert_g', 'color', 0, 255, nothing)
cv2.createTrackbar('Vert_b', 'color', 0, 255, nothing)

cv2.createTrackbar('Horz_r', 'color', 0, 255, nothing)
cv2.createTrackbar('Horz_g', 'color', 0, 255, nothing)
cv2.createTrackbar('Horz_b', 'color', 0, 255, nothing)


img4=np.zeros((256,256,3), np.uint8)

while 1:
    Vr = cv2.getTrackbarPos('Vert_r', 'color')
    Vg = cv2.getTrackbarPos('Vert_g', 'color')
    Vb = cv2.getTrackbarPos('Vert_b', 'color')
    
    
    Hr = cv2.getTrackbarPos('Horz_r', 'color')
    Hg = cv2.getTrackbarPos('Horz_g', 'color')
    Hb = cv2.getTrackbarPos('Horz_b', 'color')
    
    img4[50, :] = [Hb,Hg,Hr]
    
    img4[:,100]= [Vb,Vg,Vr]
                
    cv2.imshow('color', img4)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```

# Q5 create 250*250 rgb image with black background and make 1 vertical line and 1 horizontal line with red color and both should be controlled spereat from 2 trackbar to select the horizontal line in which row and vertical line in which column. First one “horz” and second one is “vert” (Trackbar)

```python
def nothing(x):
    pass

cv2.namedWindow("Lines")

cv2.createTrackbar('horz', 'Lines', 0, 250, nothing)
cv2.createTrackbar('vert', 'Lines', 0, 250, nothing)

img5 = np.zeros((250, 250, 3), np.uint8)

while 1:
    h = cv2.getTrackbarPos('horz', 'Lines')
    v = cv2.getTrackbarPos('vert', 'Lines')
    
    img5[h, :] = [0, 0, 255]
    img5[:, v] = [0, 0, 255]
    
    cv2.imshow('Lines', img5)
    
    img5 = np.zeros((250, 250, 3), np.uint8)
    
    k = cv2.waitKey(1)
    
    if k == ord('q'):
        break

cv2.destroyAllWindows()
```

# Q- Read from camera and make a video with 30 frames/sec

```python
import time

cap=cv2.VideoCapture(0)
if cap.isOpened() == False:
    print('Error')
    
frameSize = (640, 480)

out = cv2.VideoWriter('output_2.mp4', cv2.VideoWriter_fourcc(*'DIVX'), 30.0, frameSize)
counter = 0
tic = time.time()

while 1:
    state,frame=cap.read()
    if state==True:
        out.write(frame)
        cv2.imshow('Frame',frame)
        counter += 1
        print(time.time() - tic)
        print(f'count : {counter}')
        k=cv2.waitKey(1)
        if k== ord('q'):
            break
    else:
        break
        
tac = time.time()
print(f'time : {tac - tic}')
print(f'Frames : {counter}')
print(f'frames/sec : {counter / (tac - tic)}')
        
cap.release()
out.release()
cv2.destroyAllWindows()
```
