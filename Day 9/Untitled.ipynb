```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
src = np.array(
    [
        [7,2,3,3,8],
        [4,5,3,8,4],
        [3,3,2,8,4],
        [2,8,7,2,7],
        [5,4,4,5,4]
    ], dtype=np.float32)

ker = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
], dtype=np.float32)
```

```python
plt.imshow(src)
```

```python
out = cv2.filter2D(src, -1, ker)
```

```python
plt.imshow(out)
print(out)
```

```python
f, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(10, 10))
ax0.imshow(src, cmap='gray')
ax1.imshow(ker, cmap='gray')
ax2.imshow(out, cmap='gray')
```

```python
avg= cv2.blur(src, (9, 9))
plt.imshow(avg[:,:])
```

```python
avg1= cv2.GaussianBlur(src, (21, 21), 1)
plt.imshow(avg1)
```

```python
img = np.zeros((100, 100), dtype=np.float32)
img[:50, :] = 1
plt.imshow(img, cmap='gray')
```

```python
#callback 
def nothing(x):
    pass
```

```python
cv2.namedWindow('win')
cv2.createTrackbar('r', 'win', 1, 255, nothing)


while 1:
    size = cv2.getTrackbarPos('r', 'win')
    
    out = cv2.GaussianBlur(img, (size, size),0)
    cv2.imshow('win', out)
    k = cv2.waitKey(1)
    if k == ord('q'):
        
        break
cv2.destroyAllWindows()
```

```python
b1 = cv2.bilateralFilter(img, 30, 75, 75)
plt.imshow(b1)
```

```python
gray = plt.imread('img.jpg')
sobel_x = np.array([[1, 2, 1],
                   [0, 0, 0],
                   [-1, -2, -1]])

filtered_image = cv2.filter2D(gray, -1, sobel_x)
plt.imshow(filtered_image, cmap='gray')
```
