# Q12- Make trackbar for rgb colors with window “colors” 

```python
#callback 
def nothing(x):
    pass
```

```python
cv2.namedWindow('colors')
cv2.createTrackbar('r', 'colors', 10, 255, nothing)
cv2.createTrackbar('g', 'colors', 10, 255, nothing)
cv2.createTrackbar('b', 'colors', 10, 255, nothing)

rgbimg=np.zeros((500,500,3), np.uint8)

while 1:
    r = cv2.getTrackbarPos('r', 'colors')
    g = cv2.getTrackbarPos('g', 'colors')
    b = cv2.getTrackbarPos('b', 'colors')
    
    rgbimg[:] = [b,g,r]
    cv2.imshow('colors', rgbimg)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```
