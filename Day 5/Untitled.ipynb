```python
cap = cv2.VideoCapture('11_output.avi')

while(cap.isOpened()):
    state,frame=cap.read()
    if state == True:
        cv2.imshow('Frame', frame)
# 0 capture and hold
# to continue when pressed
        k = cv2.waitKey(2)
        if k == ord('q'):
            break
        elif k == ord('s'):
            cv2.imwrite('out.jpg', frame)
    else:
         break
cap.release()
cv2.destroyAllWindows()
```

# Track bar

```python
#callback 
def nothing(x):
    print(x)
```

```python
cv2.namedWindow('Image')
cv2.createTrackbar('val', 'Image', 10, 255, nothing)
img=np.zeros((500,500,1), np.uint8)

while 1:
    value = cv2.getTrackbarPos('val', 'Image')
    img[:] = value
    cv2.imshow('Image', img)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```
